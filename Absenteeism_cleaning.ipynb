<h1>Data Cleaning Process</h1>

## Drop ID columns

```python
df = df.drop('ID',axis= 1 )
df
```

## Grouping Reason of Absence

<li> Value 1-14 is Related to Diseases</li>
<li> Value 15-17 is Related to Plangency</li>
<li> Value  18-21 is Related to other</li>

```python
reason_columns = pd.get_dummies(df['Reason for Absence'],dtype=np.int32)
reason_columns
```

```python
#Check if there are any empty reason
reason_columns['check'] = reason_columns.sum(axis= 1)
reason_columns
```

```python
#Confirm no empty Data
reason_columns['check'].sum(axis=0)
```

```python
#drop check column
reason_columns.drop('check',axis=1)
```

```python
#Create dummies for the Reason fpr Absence Column
reason_columns = pd.get_dummies(df['Reason for Absence'],drop_first=True,dtype=np.int32)
```

```python
reason_columns
```

```python
#create column for the reason_type_1
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_1.name = 'reason type 1'
reason_type_1
```

```python
#create column for the reason_type_2
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_2.name = 'reason type 2'
reason_type_2
```

```python
#create column for the reason_type_3
reason_type_3 = reason_columns.loc[:,18:21].max(axis =1)
reason_type_3.name = 'reason type 3'
reason_type_3
```

```python
#create column for the reason_type_4
reason_type_4 = reason_columns.loc[:,22:28].max(axis=1)
reason_type_4.name = 'reason type 4'
reason_type_4
```

```python
#Drop the Column "Reason for Absence" from the original DataFrame
df = df.drop('Reason for Absence',axis=1)
df 
```

```python
#Concat the new row type columns to the DataFrame
df = pd.concat([reason_type_1,reason_type_2,reason_type_3,reason_type_4,df],axis=1)
df_save_point_1 = df 
df_save_point_1
```

## Extracting Month and Day of the week from "Date" Column

```python
#Convert the Date Formate
date_of_absence = pd.to_datetime(df.loc[:,"Date"],format='%d/%m/%Y')
date_of_absence
```

```python
#Extract the month value from the date_of_absent
month_of_absence = date_of_absence.dt.month
month_of_absence.name = 'month value'
month_of_absence
```

```python
#Extract the date of week from the date_of_absent
day_of_week_absence = date_of_absence.dt.day_of_week
day_of_week_absence.name = 'day of week'
day_of_week_absence
```

```python
#Concat the month and day of week column to the dataframe
df = pd.concat([df,month_of_absence,day_of_week_absence], axis= 1 )
df
```

```python
#Drop the Date Column and create check point
df = df.drop('Date', axis=1)
df_save_point_2 = df.copy()
df_save_point_2
```

## Categorize Education Column 
<ul>Education Column Value
<li>1 = High School</li>
<li>2 = College</li>
<li>3 = Bachelor</li>
<li>4 = Master/Doctor<li>
</ul>
<p> Objective: Categorize the Education Column into 1 which is higher education and 0 which is high school education</p>

```python
#check value
df['Education'].unique()
```

```python
#count value
df['Education'].value_counts()
```

```python
#Categorize columns Education into educate_distinct
educate_distinct = df['Education'].map({1:0,2:1,3:1,4:1})
educate_distinct
```

```python
#Replace Column
df['Education'] = educate_distinct
df
```

## Rearrange Columns and Export DataFrame

```python
df.columns
```

```python
#Rearrange Columns order
df_rendered_columns = ['reason type 1', 'reason type 2', 'reason type 3', 'reason type 4',
       'month value', 'day of week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
#Apply order to DataFrame
df = df[df_rendered_columns]
df
```

```python
#Check if Clean Completed
df_cleaned = df
df_cleaned.head()
```
